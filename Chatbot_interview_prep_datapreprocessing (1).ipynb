```python
from langdetect import detect
```

# Function for removing non-english questions

```python
def detect_english(interview_question):
  try:
    return detect(interview_question)=='en'
  except:
    return False
```

# Apply function on dataset

```python
df=df[df['interview_question'].apply(detect_english)]
```

# Resetting indexes

```python
df.reset_index(inplace=True,drop=True)
```

```python
df.info()
```

Total [90737-81655]=9082 rows removed.New shape of data 81655 rows and 4 columns.

# Remove non-readable characters.

```python
df['interview_question']=df['interview_question'].str.replace('\W',' ',regex=True)
```

```python
df.head()
```

# Remove was asked... from job title column

```python
df['job_title']=df['job_title'].str.replace('was asked...',' ')
```
